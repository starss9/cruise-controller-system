```python
# Cruise Control System Simulator in Python
!pip install control  # Install 'control'
import control as ctrl # Import 'control' immediately after installation

# Your code using 'control' goes here
import numpy as np
import matplotlib.pyplot as plt
import control as ctrl

# Car Parameters
m = 1000  # mass of the car in kg
b = 50    # damping coefficient (friction, air resistance)

# Transfer Function: G(s) = 1 / (ms + b)
num = [1]
den = [m, b]
plant = ctrl.TransferFunction(num, den)

# PID Controller Parameters
Kp = 500
Ki = 10
Kd = 50

# PID Transfer Function: Kd*s^2 + Kp*s + Ki / s
pid = ctrl.TransferFunction([Kd, Kp, Ki], [1, 0])

# Closed-loop system
open_loop = pid * plant
closed_loop = ctrl.feedback(open_loop)

# Time and Target Speed
t = np.linspace(0, 20, 1000)
target_speed = 10  # meters/second (setpoint)

# Simulate step response
t, y = ctrl.step_response(closed_loop * target_speed, t)

# Plot the response
plt.figure(figsize=(10, 5))
plt.plot(t, y, label='Speed (m/s)', linewidth=2)
plt.axhline(target_speed, color='r', linestyle='--', label='Setpoint (10 m/s)')
plt.title('Cruise Control System Response')
plt.xlabel('Time (s)')
plt.ylabel('Speed (m/s)')
plt.grid(True)
plt.legend()
plt.show()
```
